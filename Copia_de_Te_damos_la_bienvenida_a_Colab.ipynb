```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Cargar el dataset 'insurance.csv'
try:
    data = pd.read_csv('/content/insurance.csv')
except FileNotFoundError:
    print("Error: El archivo 'insurance.csv' no se encuentra.")
    print("Por favor, asegúrate de que el archivo esté en el mismo directorio de Colab.")
    exit()

# Pre-procesamiento de datos categóricos
# Codificación para 'sex' (hombre/mujer) y 'smoker' (sí/no)
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['smoker'] = le.fit_transform(data['smoker'])

# One-Hot Encoding para 'region' (múltiples categorías)
data = pd.get_dummies(data, columns=['region'], dtype=int)

# Seleccionar las características (X) y la etiqueta (y).
# Las columnas del dataframe ya han sido reordenadas
# Columnas: age, sex, bmi, children, smoker, charges, region_northeast, region_northwest, region_southeast, region_southwest
y = data['charges'].values
X_df = data.drop('charges', axis=1) # Todas las columnas excepto 'charges'
X = X_df.values
m = y.size

print(f"Número de ejemplos (m): {m}")
print(f"Número de propiedades (n): {X.shape[1]}")

def normalizarCaracteristicas(X):
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    # Evitar división por cero si la desviación estándar es 0
    sigma[sigma == 0] = 1.0e-12
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma

# Llama a normalizarCaracteristicas con los datos cargados
X_norm, mu, sigma = normalizarCaracteristicas(X)

print('Media de las características:', mu)
print('Desviación estándar de las características:', sigma)

def graficarDatos(x, y, xlabel, ylabel):
    fig = pyplot.figure()
    pyplot.plot(x, y, 'ro', ms=5, mec='k')
    pyplot.xlabel(xlabel)
    pyplot.ylabel(ylabel)
    pyplot.show()

# Ajustar las etiquetas de los gráficos a los datos médicos
xlabel1 = "Edad (normalizada)"
xlabel2 = "BMI (normalizado)"
ylabel1 = "Cargos Médicos (USD)"

# Gráficas
graficarDatos(X_norm[:, 0], y, xlabel1, ylabel1)
graficarDatos(X_norm[:, 2], y, xlabel2, ylabel1)

# Añadir unos a la matriz de X para el término de intercepción (theta_0)
X_ready = np.concatenate([np.ones((m, 1)), X_norm], axis=1)

def calcularCosto(X, y, theta):
    m = y.size
    J = (1/(2 * m)) * np.sum(np.square(np.dot(X, theta) - y))
    return J

theta_inicial = np.zeros(X_ready.shape[1])
JJ = calcularCosto(X_ready, y, theta_inicial)
print(f"Con theta inicial de ceros se obtiene un costo de: {JJ}")

def calcularDescensoGradiente(X, y, theta, alpha, numero_iteraciones):
    m = y.shape[0]
    theta = theta.copy()
    J_historico = []

    for i in range(numero_iteraciones):
        theta = theta - (alpha / m) * (np.dot(X, theta) - y).dot(X)
        J_historico.append(calcularCosto(X, y, theta))

    return theta, J_historico

num_ite = 400
alpha = 0.01
theta, J_historico = calcularDescensoGradiente(X_ready, y, theta_inicial, alpha, num_ite)

print(f"Los valores de theta calculados por Descenso de Gradiente son: {theta}")
print(f"Con un costo final de: {J_historico[-1]}")

# Graficar la convergencia del costo
pyplot.plot(np.arange(len(J_historico)), J_historico, lw=2)
pyplot.xlabel('Número de iteraciones')
pyplot.ylabel('Costo J')
pyplot.title('Convergencia del Costo con Descenso de Gradiente')
pyplot.show()

# Predicciones con Descenso de Gradiente
# Ejemplo: Un hombre de 30 años, BMI 25, 1 hijo, no fumador, de la región 'southwest'.
ejemplo_sin_norm = np.array([30, 0, 25.0, 1, 0, 0, 0, 1, 0])
# Normalizamos los valores
ejemplo_num_norm = (ejemplo_sin_norm - mu) / sigma
# Unimos el 1 para el término de intercepción
ejemplo_final_completo = np.concatenate([np.ones(1), ejemplo_num_norm])
y_pred_dg = np.dot(ejemplo_final_completo, theta)
print(f"Predicción de cargos (usando Descenso de Gradiente): ${y_pred_dg:,.2f}")

# Utilizando la Ecuación de la Normal
def calcularThethaEcuacionNormal(X, y):
    theta = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
    return theta

# Se usa la matriz X sin normalizar, pero con la columna de unos
X_ren = np.concatenate([np.ones((m, 1)), X], axis=1)
theta_normal = calcularThethaEcuacionNormal(X_ren, y)
print(f"Los valores de theta calculados por la Ecuación Normal son: {theta_normal}")

# Predicciones con Ecuación Normal
# Se usan los datos sin normalizar, igual que en la normalización para el cálculo del theta.
ejemplo_sin_norm_ecnormal = np.array([1, 30, 0, 25.0, 1, 0, 0, 0, 1, 0])
y_pred_normal = np.dot(ejemplo_sin_norm_ecnormal, theta_normal)
print(f"Predicción de cargos (usando Ecuación Normal): ${y_pred_normal:,.2f}")
```
